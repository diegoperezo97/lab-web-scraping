#### Number of Titles that have changed in the United States Code since its last release point 

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'lxml')
````

````python
# Your code
titles_changed = soup.select('div[class="usctitle"]')
titles_changed_clean = [title_changed.text.strip().replace('\n\n         ', '').replace(' ٭', '') \
                        for title_changed in titles_changed]
titles_changed_clean = titles_changed_clean[2:]
print('Number of Titles that have changed in the United States Code since its last release point:', \
      len(titles_changed_clean))
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'lxml')
````

````python
# Your code
languages_html = []
n = 1

while n < 11:
    x = soup.select(f'div[class="central-featured-lang lang{n}"]')
    languages_html.append(x[0])
    n = n + 1
    
languages = [language.text.strip() for language in languages_html]
languages_clean = [re.sub(r'\n.*', '', language) for language in languages]
languages_clean
````
